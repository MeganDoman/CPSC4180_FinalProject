# Read Popular Vote Data

```python
votesdf = pd.read_csv('PopularVote.csv')
votesdf
```

```python
votesdf = votesdf.set_index('stateid')
```

```python
votesdf
```

```python
votesindex = votesdf.index
votesindex
```

```python
tweetdf = tweetdf.transpose()
tweetdf
```

```python
tweetdf.columns = tweetdf.columns.strftime('%Y-%m-%d')
tweetdf
```

```python
tweetindex = tweetdf.index
tweetindex
```

```python
combinedindex = set(tweetindex) & set(votesindex)
print(combinedindex)
```

```python
votesnumdf = votesdf.loc[combinedindex]
votesnumdf = votesnumdf.filter(["called", "dem_percent","rep_percent","other_percent"], axis=1)
votesnumdf
```

```python
combinedf = pd.merge(votesnumdf, tweetdf, left_index=True, right_index=True, how='inner')
combinedf
```

```python

votecols = ['dem_percent', 'rep_percent', "other_percent"]
for col in votecols:
    combinedf[col] = combinedf[col].str.rstrip('%').astype('float')
    combinedf[col] = combinedf[col]/100
combinedf.head()
```

# plot the data as time series / line graph

```python
rep = combinedf[combinedf["called"] == "R"]
rep_ts = rep.iloc[:, 4:]
dem = combinedf[combinedf["called"] == "D"]
dem_ts = dem.iloc[:, 4:]
dem_ts.loc['mean'] = dem_ts.mean()
rep_ts.loc['mean'] = rep_ts.mean()
dem_mean = dem_ts.loc["mean"]
rep_mean = rep_ts.loc["mean"]
```

```python
dem_ts.columns
```

```python
# gca stands for 'get current axis'
ax = plt.gca()


#Goal: two lines, time series, average sentiment of states w/ called = R and one line for states called with D per day
dem_mean.plot(kind='line', color = 'blue', ax=ax, label = "Democrat called")
rep_mean.plot(kind='line', color = 'red', ax = ax, label  = "Republican called")
ax.set_ylim(ymin=-1, ymax=1)
ax.legend()
plt.xticks(rotation = 45)
plt.title("Average Daily Twitter Sentiment During Voting \n by State Popular Vote")
plt.ylabel("Average Twitter Sentiment")
plt.xlabel("Date")
plt.show()
```

# smooth the data

```python
combined_array = np.array([dem_mean, rep_mean])
madf = pd.DataFrame(combined_array, columns = dem_ts.columns, index = ['dem', 'rep'])
madf = madf.transpose()
madf
```

```python
madf.columns
```

```python
filter_len = 15
dem_moving_avg = np.convolve(madf['dem'], np.ones((filter_len)), mode='same')
dem_moving_avg /= filter_len
rep_moving_avg = np.convolve(madf['rep'], np.ones((filter_len)), mode='same')
rep_moving_avg /= filter_len
```

```python
ax = plt.gca()
plt.plot(madf.index.values, dem_moving_avg, color = 'blue', label = "Democrat called")
plt.plot(madf.index.values, rep_moving_avg, color = 'red', label  = "Republican called")
plt.ylim(ymin=-1, ymax=1)
plt.legend()
ax.set_xticks(np.arange(0, len(dem_moving_avg)+1, 28))
plt.xticks(rotation = 90)
plt.title("Average Daily Twitter Sentiment During Voting \n by State Popular Vote")
plt.ylabel("Average Twitter Sentiment")
plt.xlabel("Date")
#plt.ylim(ymin=-1, ymax=1)
plt.show()
```

```python
from scipy.signal import medfilt
ax = plt.gca()
dem_medfilt = medfilt(madf['dem'], filter_len)
rep_medfilt = medfilt(madf['rep'], filter_len)
plt.plot(madf.index.values, dem_medfilt, color = 'blue', label = "Democrat called")
plt.plot(madf.index.values, rep_medfilt, color = 'red', label  = "Republican called")
ax.set_ylim(ymin=-1, ymax=1)
ax.legend()
ax.set_xticks(np.arange(0, len(dem_moving_avg)+1, 28))
plt.xticks(rotation = 90)
plt.title("Average Daily Twitter Sentiment During Voting \n by State Popular Vote")
plt.ylabel("Average Twitter Sentiment")
plt.xlabel("Date")
plt.ylim(ymin=-1, ymax=1)
plt.show()
```
