# Read Popular Vote Data

```python
votesdf = pd.read_csv('PopularVote.csv')
votesdf
```

```python
votesdf = votesdf.set_index('stateid')
```

```python
votesdf
```

```python
votesindex = votesdf.index
votesindex
```

```python
tweetdf = tweetdf.transpose()
tweetdf
```

```python
tweetindex = tweetdf.index
tweetindex
```

```python
combinedindex = set(tweetindex) & set(votesindex)
combinedindex
```

```python
votesnumdf = votesdf.loc[combinedindex]
votesnumdf = votesnumdf.filter(["called", "dem_votes","rep_votes","other_votes"], axis=1)
votesnumdf
```

```python
combinedf = pd.merge(votesnumdf, tweetdf, left_index=True, right_index=True, how='inner')
combinedf
```

```python
myState = "CA"
vote_values = []
for row in votesdf.loc[myState]:
    vote_values.append(row)
vote_count = []
for val in vote_values[3:6]:
    num = val.replace(",","")
    vote_count.append(int(num))
vote_party = ["Democrat", "Republican", "Other"]
```

```python
print(vote_count)
print(vote_party)
```

```python
plt.pie(x=vote_count, labels=vote_party)
```
